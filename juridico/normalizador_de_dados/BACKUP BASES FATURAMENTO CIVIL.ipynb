```python
# --- 1. PREPARAÇÃO ---
data_hoje = datetime.now().strftime("%Y-%m-%d")
nome_arquivo = f"Backup_base_civil_consolidado_{data_hoje}.xlsx"

# Caminhos
caminho_temporario = f"/tmp/{nome_arquivo}"
caminho_final = f"/Volumes/databox/juridico_comum/arquivos/financeiro/Faturamento_Civel/BACKUP_ARQUIVOS_CONSOLIDADOS/{nome_arquivo}"

print(f"Iniciando backup de: {nome_arquivo}")

# Se o seu df_pandas já existe na memória, comente a linha abaixo.
# Se não, deixe ela ativa para criar o DF a partir da tabela Delta.
df_pandas = spark.table("databox.juridico_comum.br_Consolidado_Honorarios").toPandas()

# --- 3. GERAR EXCEL (SALVAR NO /tmp) ---
print("Gerando Excel temporário...")
df_pandas.to_excel(caminho_temporario, index=False)

# --- 4. MOVER PARA O DBFS (DOWNLOAD) ---
print("Movendo para pasta de download...")
# Garante que a pasta existe
os.makedirs(os.path.dirname(caminho_final), exist_ok=True)

# Move o arquivo
shutil.move(caminho_temporario, caminho_final)

print(f"✅ Sucesso! Arquivo salvo em: {caminho_final}")
```
