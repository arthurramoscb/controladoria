```python
df_dados_1 = spark.sql('''

select SUM(ESTOQUE) AS ESTOQUE_QTD
,SUM(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN ESTOQUE ELSE 0 END) AS ESTOQUE_NOVO
,SUM(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN ESTOQUE ELSE 0 END) AS ESTOQUE_LEGADO
 ,sum(coalesce(PROVISAO_TOTAL_M,0)) / 1000 AS SALDO_PROVISAO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN coalesce(PROVISAO_TOTAL_M,0) ELSE 0 END) / 1000 AS SALDO_PROVISAO_NOVO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN coalesce(PROVISAO_TOTAL_M,0) ELSE 0 END) / 1000 AS SALDO_PROVISAO_LEGADO
 ,sum(coalesce(CORRECAO_M,0)) / 1000  AS MOVIMENTACAO_CORRECAO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN coalesce(CORRECAO_M,0) ELSE 0 END) / 1000  AS MOVIMENTACAO_CORRECAO_NOVO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN coalesce(CORRECAO_M,0) ELSE 0 END) / 1000  AS MOVIMENTACAO_CORRECAO_LEGADO
 ,sum(coalesce(CORRECAO_M_1,0)) / 1000  AS SALDO_CORRECAO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN coalesce(CORRECAO_M_1,0) ELSE 0 END) / 1000  AS SALDO_CORRECAO_NOVO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN coalesce(CORRECAO_M_1,0) ELSE 0 END) / 1000  AS SALDO_CORRECAO_LEGADO
 ,sum(coalesce(PROVISAO_TOTAL_PASSIVO_M, 0)) / 1000  as SALDO_PROVISAO_TOTAL
 ,sum(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / 1000  as SALDO_PROVISAO_TOTAL_NOVO
 ,sum(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / 1000  as SALDO_PROVISAO_TOTAL_LEGADO
 ,(sum(coalesce(PROVISAO_TOTAL_PASSIVO_M, 0)) / count(estoque)) / 1000  AS TKM_MEDIO
  ,(sum(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / count(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN estoque ELSE 0 END)) / 1000  AS TKM_MEDIO_NOVO
   ,(sum(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / count(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN estoque ELSE 0 END)) / 1000  AS TKM_MEDIO_LEGADO
 ,SUM(NOVOS) AS ENTRADAS_QTD
 ,SUM(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN NOVOS ELSE 0 END) AS ENTRADAS_QTD_NOVO
 ,SUM(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN NOVOS ELSE 0 END) AS ENTRADAS_QTD_LEGADO
 ,SUM(CASE WHEN NOVOS = 1 THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END)  / 1000  AS ENTRADAS_VLR
 ,SUM(CASE WHEN NOVOS = 1 AND NOVO_X_LEGADO = 'NOVO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END)  / 1000  AS ENTRADAS_VLR_NOVO
 ,SUM(CASE WHEN NOVOS = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END)  / 1000  AS ENTRADAS_VLR_LEGADO
 ,(SUM(CASE WHEN NOVOS = 1 THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(NOVOS)) / 1000  AS TKM_ENTRADAS
 ,(SUM(CASE WHEN NOVOS = 1 AND NOVO_X_LEGADO = 'NOVO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN  NOVOS ELSE 0 END)) / 1000  AS TKM_ENTRADAS_NOVO
 ,(SUM(CASE WHEN NOVOS = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN  NOVOS ELSE 0 END)) / 1000  AS TKM_ENTRADAS_LEGADO
 ,SUM(ENCERRADOS) AS ENCERRADOS_QTD
 ,SUM(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN ENCERRADOS ELSE 0 END) AS ENCERRADOS_QTD_NOVO
 ,SUM(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN ENCERRADOS ELSE 0 END) AS ENCERRADOS_QTD_LEGADO
 ,SUM(CASE WHEN ENCERRADOS = 1 THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / 1000  AS TOTAL_PAGAMENTOS
 ,SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'NOVO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / 1000  AS TOTAL_PAGAMENTOS_NOVO
 ,SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / 1000  AS TOTAL_PAGAMENTOS_LEGADO
 ,(SUM(CASE WHEN ENCERRADOS = 1 THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / count(ENCERRADOS)) / 1000  AS TKM_MEDIO_ENCERRADOS
 ,(SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'NOVO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / count(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN ENCERRADOS ELSE 0 END)) / 1000  AS TKM_MEDIO_ENCERRADOS_NOVO
 ,(SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / count(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN ENCERRADOS ELSE 0 END)) / 1000  AS TKM_MEDIO_ENCERRADOS_LEGADO
 ,SUM(CASE WHEN ESTOQUE = 1 AND coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) <> 0 THEN 1 ELSE 0 END) AS ESTOQUE_COM_PROVISAO
 ,SUM(CASE WHEN ESTOQUE = 1 AND coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) <> 0 AND NOVO_X_LEGADO = 'NOVO' THEN 1 ELSE 0 END) AS ESTOQUE_COM_PROVISAO_NOVO
 ,SUM(CASE WHEN ESTOQUE = 1 AND coalesce(PROVISAO_TOTAL_PASSIVO_M, 0) <> 0 AND NOVO_X_LEGADO = 'LEGADO' THEN 1 ELSE 0 END) AS ESTOQUE_COM_PROVISAO_LEGADO
 ,(SUM(CASE WHEN ESTOQUE = 1 THEN COALESCE(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(ESTOQUE)) / 1000  AS TKM_MEDIO_COM_PROV
,(SUM(CASE WHEN ESTOQUE = 1 AND NOVO_X_LEGADO = 'NOVO' THEN COALESCE(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'NOVO' THEN ESTOQUE ELSE 0 END)) / 1000  AS TKM_MEDIO_COM_PROV_NOVO
,(SUM(CASE WHEN ESTOQUE = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN COALESCE(PROVISAO_TOTAL_PASSIVO_M, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'LEGADO' THEN ESTOQUE ELSE 0 END)) / 1000  AS TKM_MEDIO_COM_PROV_LEGADO

--  ,(SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'NOVO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'NOVO' AND ENCERRADOS = 1 THEN 1 END)) /1000 AS TKM_MEDIO_NOVO
--  ,(SUM(CASE WHEN ENCERRADOS = 1 AND NOVO_X_LEGADO = 'LEGADO' THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) / COUNT(CASE WHEN NOVO_X_LEGADO = 'LEGADO' AND ENCERRADOS = 1 THEN 1 END)) /1000 AS TKM_MEDIO_LEGADO
 ,mes_fech 
 from df_base_tabelao 
group by mes_fech
''')

df_dados_2 = spark.sql('''
    select COUNT(ENCERRADOS) AS SEM_ONUS_QTD, mes_fech from df_base_tabelao
    where MOTIVO_ENC_AGRP = 'SEM ONUS'
    GROUP BY mes_fech
''')

df_dados_3 = spark.sql('''
    select COUNT(ENCERRADOS) AS CONDENACAO_ACORDO_QTD, mes_fech from df_base_tabelao
    where MOTIVO_ENC_AGRP <> 'SEM ONUS'
    GROUP BY mes_fech
''')


df_dados_4 = spark.sql('''
    select COUNT(ENCERRADOS) AS CONDENACAO_QTD, mes_fech from df_base_tabelao
    where MOTIVO_ENC_AGRP = 'CONDENACAO'
    GROUP BY mes_fech
''')

df_dados_5 = spark.sql('''
    select COUNT(ENCERRADOS) AS ACORDO_QTD, mes_fech from df_base_tabelao
    where MOTIVO_ENC_AGRP = 'ACORDO'
    GROUP BY mes_fech
''')

df_dados_1.createOrReplaceTempView('df_dados_1')
df_dados_2.createOrReplaceTempView('df_dados_2')
df_dados_3.createOrReplaceTempView('df_dados_3') 
df_dados_4.createOrReplaceTempView('df_dados_4')
df_dados_5.createOrReplaceTempView('df_dados_5')
```

```python
base_resumo_proprio_terceiro = spark.sql('''
    SELECT
    SUM(NOVOS) AS TOTAL_ENTRADAS
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL NOT IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN NOVOS ELSE 0 END) AS TOTAL_ENTRADAS_PROPRIO
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN NOVOS ELSE 0 END) AS TOTAL_ENTRADAS_TERCEIRO
    ,SUM(ENCERRADOS) AS TOTAL_ENCERRADOS
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL NOT IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN ENCERRADOS ELSE 0 END) AS TOTAL_ENCERRADOS_PROPRIO
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN ENCERRADOS ELSE 0 END) AS TOTAL_ENCERRADOS_TERCEIRO
    ,SUM(COALESCE(TOTAL_PAGAMENTOS, 0)) AS TOTAL_PAGAMENTOS
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL NOT IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) AS TOTAL_PAGAMENTOS_PROPRIO
    ,SUM(CASE WHEN NATUREZA_OPERACIONAL IN ('TERCEIRO INSOLVENTE','TERCEIRO INSOLVENTE') THEN COALESCE(TOTAL_PAGAMENTOS, 0) ELSE 0 END) AS TOTAL_PAGAMENTOS_TERCEIRO
    ,mes_fech
    from df_base_tabelao 
    group by mes_fech
    order by mes_fech desc
''')


```

```python
df_dados_final = spark.sql('''

select A.*,
    B.SEM_ONUS_QTD,
    C.CONDENACAO_ACORDO_QTD,
    D.CONDENACAO_QTD,
    E.ACORDO_QTD

from df_dados_1 A 
left join df_dados_2 B
ON A.mes_fech = B.mes_fech
left join df_dados_3 C
ON A.mes_fech = C.mes_fech
left join df_dados_4 D
ON A.mes_fech = D.mes_fech
left join df_dados_5 E
ON A.mes_fech = E.mes_fech
''')

```
